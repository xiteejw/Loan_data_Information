```python
# Handling Missing Values
# Decide how to handle missing values, whether by removing rows, filling with statistical measures (mean, median etc) 
# or applying more complex imputation methods.
```

```python
for col in df_loan.columns:
    if df_loan[col].isnull().any():
        if df_loan[col].dtype in ['int64', 'float64']:  # Check if column is numeric
            df_loan[col] = df_loan[col].fillna(df_loan[col].median())
        else:  # For non-numeric columns
            df_loan[col] = df_loan[col].fillna(df_loan[col].mode()[0])

# Verify if there are any more missing values:
print("\nMissing values after imputation:")
print(df_loan.isnull().sum())
```

```python
# 2)Is there a relationship between the borrower's location (zip code or state) and their loan default rate?
```

```python
df_loan['defaulted'] = np.where(df_loan['loan_status'] == 'Charged Off', 1, 0)

# Group by state and calculate total loans and defaulted loans
default_rate_by_state = df_loan.groupby('addr_state' , observed= False).agg(
    total_loans=('defaulted', 'count'),
    defaulted_loans=('defaulted', 'sum')
).reset_index()

# Calculate default rate
default_rate_by_state['default_rate'] = (default_rate_by_state['defaulted_loans'] / default_rate_by_state['total_loans']) * 100

# Sort by default rate
default_rate_by_state = default_rate_by_state.sort_values(by='default_rate', ascending=False)
default_rate_by_state = default_rate_by_state.round(2)
print(default_rate_by_state)
```

```python
# Sort the data by 'default_rate' in descending order
default_rate_by_state = default_rate_by_state.sort_values(by='default_rate', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(
    y='default_rate', 
    x='addr_state', 
    data=default_rate_by_state, 
    palette='viridis', 
    order=default_rate_by_state['addr_state']  # Ensures correct order
)
plt.title('Default Rate by State')
plt.xlabel('State')
plt.ylabel('Default Rate (%)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

Observation :
* Nebraska has the higest Default rate
* Iowa has the lowest default rate

```python
# 3)How does the length of the loan term (e.g. 36 months, 60 months) impact the interest rate and default rate? 
```

```python
df_loan['defaulted'] = np.where(df_loan['loan_status'] == 'Charged Off', 1, 0)

# Group by loan term (e.g., 36 or 60 months) and calculate average interest rate and default rate
loan_term_stats = df_loan.groupby('term' ).agg(
    average_interest_rate=('int_rate', 'mean'),
    total_loans=('defaulted', 'count'),
    defaulted_loans=('defaulted', 'sum')
).reset_index()

# Calculate the default rate
loan_term_stats['default_rate'] = (loan_term_stats['defaulted_loans'] / loan_term_stats['total_loans']) * 100
loan_term_stats = loan_term_stats.round(2)
# Display the results
print(loan_term_stats)
```

```python
# Set up a figure
plt.figure(figsize=(12, 6))

# Plot average interest rate by loan term
plt.subplot(1, 2, 1)
sns.barplot(x='term', y='average_interest_rate', data=loan_term_stats, palette='coolwarm')
plt.title('Average Interest Rate by Loan Term')
plt.xlabel('Loan Term (Months)')
plt.ylabel('Average Interest Rate (%)')

# Plot default rate by loan term
plt.subplot(1, 2, 2)
sns.barplot(x='term', y='defaulted_loans', data=loan_term_stats, palette='coolwarm')
plt.title('Default Rate by Loan Term')
plt.xlabel('Loan Term (Months)')
plt.ylabel('Defaulted Loan')

plt.tight_layout()
plt.show()
```

Observation:
* The analysis reveals a clear relationship between loan term and financial outcomes. Longer loan terms are associated with higher interest rates and increased default rates, suggesting a greater financial risk for borrowers.

```python
# 4)What is the percentage of borrowers who are repeat borrowers (i.e., have taken out multiple loans)?
```

```python
# Assuming 'df_loan' is your DataFrame and it has a column 'member_id' (or similar) for unique borrowers
# Grouping by the borrower ID and counting the number of loans per borrower
borrower_loan_counts = df_loan.groupby('member_id').size()

# Identifying repeat borrowers (borrowers who have more than 1 loan)
repeat_borrowers = borrower_loan_counts[borrower_loan_counts > 1]

# Calculate the total number of unique borrowers
total_borrowers = borrower_loan_counts.shape[0]

# Calculate the number of repeat borrowers
total_repeat_borrowers = repeat_borrowers.shape[0]

# Calculate the percentage of repeat borrowers
percentage_repeat_borrowers = (total_repeat_borrowers / total_borrowers) * 100

print(f"Percentage of repeat borrowers: {percentage_repeat_borrowers:.2f}%")
```
